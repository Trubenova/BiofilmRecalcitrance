```python
import sys
import os
import time
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import itertools

%matplotlib widget
from mpl_toolkits.mplot3d import Axes3D

# importing required libraries
```

```python
def Get_right_subset(inputs, results, cost=0.5,T1=24):
    subset1=inputs[inputs[:,0]==cost]
    subset2=subset1[subset1[:,3]==T1]
    
    r_subset1=results[inputs[:,0]==cost]
    r_subset2=r_subset1[subset1[:,3]==T1]
    
    return(subset2, r_subset2)

def Plot_3D_graph(inputs, results, which, title=''):
    # this function takes selected inputs and outputs and creates 3D figures for all outputs, for 2 inputs fixed and 2 inputs variable
    ## inputs = [cost, release, adhesion, T1]
    ## results =[plaG,plaR, bioG, bioR]
    X=inputs[:,1]
    X=np.log10(X)
    Y=inputs[:,2]
    Y=np.log10(Y)
    Z=results[:,which]
    Z=np.log10(Z)
    X=np.reshape(X, [8,8])
    Y=np.reshape(Y, [8,8])
    Z=np.reshape(Z, [8,8])
    
    
    figg = plt.figure()
    ax = figg.add_subplot(111, projection='3d')

    # creating the plot
    #ax = plt.axes(projection='3d')
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
    ax.set_title(title);

    # setting title and labels
    ax.set_xlabel('release')
    ax.set_ylabel('adesion')
    ax.set_zlabel('number')

    # displaying the plot
    plt.show()
    print (Z)
#    print (X,Y,Z)
```

```python
inputs=np.load('E2_inputs.npy')
results=np.load('E2_results.npy')
dynamics=np.load('test.npy')
#name='E2_C'+str(cost)+'Rho'+str(release)+'Alpha'+str(adhesion)+'T1'+str(T1)
```

```python
subset2, r_subset2=Get_right_subset(inputs, results, cost=0.2,T1=24)    
r_subset2=r_subset2+1
Plot_3D_graph(subset2, r_subset2, 0,  title='pla green')
Plot_3D_graph(subset2, r_subset2, 1,  title='pla red')
Plot_3D_graph(subset2, r_subset2, 2,  title='bio green')
Plot_3D_graph(subset2, r_subset2, 3,  title='bio red')

ratios_pla=r_subset2[:,0]/r_subset2[:,1]
ratios_bio=r_subset2[:,2]/r_subset2[:,3]
#ratios_pla=np.log10(ratios_pla)
#ratios_bio=np.log10(ratios_bio)

ratios=np.stack((ratios_pla,ratios_bio), axis=1)


#ratios=np.log10(ratios)
print (np.shape(ratios))
Plot_3D_graph(subset2, ratios, 0,  title='ratio pla')
Plot_3D_graph(subset2, ratios, 1,  title='ratio bio')
```
